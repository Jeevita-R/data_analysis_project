```python
import pandas as pd
df=pd.read_csv('sales_data.csv',encoding='ISO-8859-1')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df[df['Amount'].isnull()]
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.rename(columns={'Marital_Status':'Marital'},inplace=True)
df.head()
```

```python
df['Marital']=df['Marital'].map({1:'single',0:'married'})
df.head()
```

```python
df['Gender']=df['Gender'].map({'F':'Female','M':'Male'})
df.head()
```

```python
import seaborn as sns
a=sns.countplot(x='Gender',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
a=df.groupby('Gender')['Amount'].sum()
a=pd.DataFrame(a)
b=sns.barplot(x='Gender',y='Amount',data=a,hue='Gender')
for i in b.containers:
  b.bar_label(i)
```

```python
a=sns.countplot(x='Age Group',data=df,hue='Gender')
for i in a.containers:
  a.bar_label(i)
```

```python
a=df.groupby('Age Group')['Amount'].sum()
a=pd.DataFrame(a)
b=sns.barplot(x='Age Group',y='Amount',data=a)
for i in b.containers:
  b.bar_label(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(19,5))
a=sns.countplot(x='State',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
a=df.groupby('State')['Amount'].sum()
a=pd.DataFrame(a)
plt.figure(figsize=(19,5))
b=sns.barplot(x='State',y='Amount',data=a)
for i in b.containers:
  b.bar_label(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(19,5))
a=sns.countplot(x='Occupation',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
a=df.groupby('Occupation')['Amount'].sum()
a=pd.DataFrame(a)
plt.figure(figsize=(19,5))
b=sns.barplot(x='Occupation',y='Amount',data=a)
for i in b.containers:
  b.bar_label(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(19,5))
a=sns.countplot(x='Product_Category',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
a=df.groupby('Product_Category')['Amount'].sum()
a=pd.DataFrame(a)
plt.figure(figsize=(19,5))
b=sns.barplot(x='Product_Category',y='Amount',data=a)
for i in b.containers:
  b.bar_label(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(19,5))
a=sns.countplot(x='Marital',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
a=df.groupby('Marital')['Amount'].sum()
a=pd.DataFrame(a)
plt.figure(figsize=(19,5))
b=sns.barplot(x='Marital',y='Amount',data=a)
for i in b.containers:
  b.bar_label(i)
```
